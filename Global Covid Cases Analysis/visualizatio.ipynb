```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("Ready_For_Visualisation.csv")

top_countries = df.sort_values('Total_confirmed_cases', ascending=False).head(10)

top_countries.head(10)
```

```python
top_countries.shape
```

```python
#Bar graph Countries vs Recovered Percentage

plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Recovered_percentage', data=top_countries, palette='viridis')
plt.title("Top 10 Countries by Total Confirmed COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Recovered Percentage")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar graph - Country vs Death percentage  

plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Death_percentage', data=top_countries, palette='viridis')
plt.title("Country vs Death Percentage")
plt.xlabel("Countries")
plt.ylabel("Death Percentage")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_countries, x='Total_confirmed_cases', y='Recovered_covid', hue='Country/Region', palette='tab10', s=100)

plt.title('Scatter Plot: Total Confirmed Cases vs Recovered Cases by Province/State')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Recovered Cases')

plt.legend(title='Country/Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Conclusion of this analysis

The data shows that among the top 10 countries with the most COVID-19 cases, death rates vary a lot. Brazil and Russia have higher death percentages, while South Korea has managed to keep it very low.

Despite the differences in death rates, almost all these countries have very high recovery rates, with over 95% of infected people recovering.

Having a high number of confirmed cases doesn’t necessarily mean a country will have a high death rate. This suggests that how a country handles the pandemic matters a great deal.

Countries like South Korea, which kept deaths low and recoveries high, likely benefited from early testing, strong healthcare, and effective government measures.

Overall, while most people recover from COVID-19, the risk of death depends on factors like healthcare quality and public health policies.

Understanding these patterns can help guide better responses for any future outbreaks, aiming to save more lives and support recovery.
