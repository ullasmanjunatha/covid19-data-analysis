```python
import pandas as pd

df = pd.read_csv('RAW_global_confirmed_cases.csv')

df.head(10)
```

```python
#Removing useless columns
useless_column = ['Province/State', 'Lat', 'Long']

df1 = df.drop(useless_column, axis=1)

df1.head(10)


```

```python
#Shape of the DataFrame
df1.shape
```

```python
#Aggregated

df2 = df1.groupby('Country/Region').sum().reset_index()

df2.shape
```

I have successfully Removed columns like Province/State, Lat, Long and combined all the provinces that belonged to same country.

```python
df2['Total_confirmed_cases'] = df2.loc[:, df2.columns != 'Country/Region'].sum(axis=1)

df2.head(10)
```

```python
# Keeping only data required for the analysis 

df3 = df2[['Country/Region', 'Total_confirmed_cases']]

df3.to_csv("Cleaned_global_confirmed_cases.csv", index=False)
```

I have cleaned the data and exported into new file "Cleaned_global_confirmed_cases.csv"
