```python
import pandas as pd

df = pd.read_csv("Raw_global_deaths.csv")

df.head(10)
```

```python
df.shape
```

```python
#Deleting useless columns
useless_columns = ['Province/State', 'Lat', 'Long']

df1 = df.drop(useless_columns,axis=1)

df1.head()
```

```python
#aggregating countries

df2 = df1.groupby('Country/Region').sum().reset_index()

df2.shape
```

```python
#Sum of all deaths for each countries

df2['Total_deaths'] = df2.loc[:, df2.columns != 'Country/Region' ].sum(axis=1)

df2.head()
```

```python
# keeping only required columns and exporting into new csv file

df3 = df2[['Country/Region', 'Total_deaths']]

df3.to_csv('Cleaned_global_deaths.csv', index=False)
```

I have removed unwanted rows in the file, calculated total deaths and exported into new file.
