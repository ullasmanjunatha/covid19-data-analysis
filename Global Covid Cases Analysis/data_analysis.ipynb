```python
import pandas as pd

cdf1 = pd.read_csv("Cleaned_global_confirmed_cases.csv")
cdf2 = pd.read_csv("Cleaned_global_deaths.csv")

merged_df = pd.merge(cdf1, cdf2, on='Country/Region', how='inner')

merged_df.head(10)
```

```python
# Calculating Recovered from covid

merged_df['Recovered_covid'] = merged_df['Total_confirmed_cases'] - merged_df['Total_deaths']

merged_df.head(10)
```

```python
#recovered percentage & death percentage

merged_df['Recovered_percentage'] = (merged_df['Recovered_covid'] / merged_df['Total_confirmed_cases']) *100

merged_df['Death_percentage'] = (merged_df['Total_deaths'] / merged_df['Total_confirmed_cases']) * 100

merged_df.head(10)

```

```python
# Exporting data to new file

merged_df.to_csv("Ready_For_Visualisation.csv", index=False)
```
