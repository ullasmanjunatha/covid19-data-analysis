```python
import pandas as pd

df = pd.read_csv("RAW_us_deaths.csv")

df.head(10)
```

```python
df.shape
```

```python
#remove useless columns

useless_columns = ['Admin2', 'UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Country_Region', 'Lat', 'Long_']

df1 = df.drop(useless_columns, axis=1, errors='ignore')

df1.head(10)

```

```python
#Aggregate countries

df1 = df1.groupby('Province_State').sum().reset_index()

df1.head(10)
```

```python
df1.shape
```

```python
#ADD total deaths

columns_to_add = df1.columns.difference(['Province_State', 'Combined_Key', 'Population'])

df1['Total_Deaths'] = df1[columns_to_add].sum(axis=1)

df1.head(10)
```

```python
#Keep required data

df2 = df1[['Province_State', 'Total_Deaths']]

df2.head(10)
```

```python
#Export to new file

df2.to_csv("Cleaned_us_deaths.csv", index=False)
```
