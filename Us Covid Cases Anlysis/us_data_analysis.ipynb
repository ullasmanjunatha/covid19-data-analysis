```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Merging two cleanded files to analyze

```python
df1 = pd.read_csv("Cleaned_us_confirmed_cases.csv")
df2 = pd.read_csv("Cleaned_us_deaths.csv")

merged_us = pd.merge(df1, df2, on='Province_State', how='inner')

merged_us.head(10)
```

```python
# Calculating recovered cases, 

merged_us['Recovered_Cases'] = merged_us['Total_Confirmed_Cases'] - merged_us['Total_Deaths']


merged_us.head(10)
```

```python
#Calculating recovered case and deaths percentage

merged_us['Recovered_Percentage'] = (merged_us['Recovered_Cases'] / merged_us['Total_Confirmed_Cases']) *100

merged_us['Death_Percentage'] = (merged_us['Total_Deaths']/ merged_us['Total_Confirmed_Cases']) * 100

merged_us.head(10)
```

```python
# Top 10 Province with most number of covid cases

top_province = merged_us.sort_values('Total_Confirmed_Cases', ascending=False).head(10)


top_province.shape
```

```python
# Export data for visualization

top_province.to_csv("Visualization_Ready.csv", index=False)
```
