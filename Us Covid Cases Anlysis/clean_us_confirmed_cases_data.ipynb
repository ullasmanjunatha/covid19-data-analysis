```python
import pandas as pd

df = pd.read_csv("RAW_us_Confirmed_cases.csv")

df.head(10)
```

```python
df.shape
```

```python
#Aggrigating Province States
df1 = df.groupby('Province_State').sum().reset_index()
df1.head(10)
```

```python
df1.shape
```

```python
useless_column = ['Admin2', 'UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Country_Region', 'Lat', 'Long_']

df2 = df1.drop(useless_column, axis=1)

df2.head(10)
```

```python
df2.shape
```

```python
# Sum of total confirmed cases

columns_to_sum = df2.columns.difference(['Province_State', 'Combined_Key'])

df2['Total_Confirmed_Cases'] = df2[columns_to_sum].sum(axis=1)

df2.head(10)

```

```python
#keep only required data 

df3 = df2[['Province_State', 'Total_Confirmed_Cases']]

df3.head(10)
```

```python
#Export to new file

df3.to_csv("Cleaned_us_confirmed_cases.csv", index=False)
```
