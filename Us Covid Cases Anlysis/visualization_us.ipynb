```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Visualization_Ready.csv")

df.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x= 'Province_State', y='Recovered_Cases', data=df, palette='viridis')
plt.title("Top 10 Province with highest confirmed cases vs recovered cases")
plt.xlabel("Province/State")
plt.ylabel("Pecovered Percentage")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# confirmed cases vs death 
plt.figure(figsize=(12,6))
sns.barplot(x= 'Province_State', y='Death_Percentage', data=df, palette='viridis')
plt.title("Top 10 Province with highest confirmed cases vs Death Percentage")
plt.xlabel("Province/State")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Total_Confirmed_Cases', y='Recovered_Cases', hue='Province_State', palette='tab10', s=100)

plt.title('Scatter Plot: Total Confirmed Cases vs Recovered Cases by Province/State')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Recovered Cases')

plt.legend(title='Province_State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Data analysis

California has the highest number of confirmed and recovered COVID-19 cases among all provinces/states .

The scatter plot shows a strong positive relationship between confirmed cases and recovered cases across top states, indicating effective recovery rates despite large outbreaks.

Total deaths vary across states, with New York, Pennsylvania, and Michigan recording the highest death counts among the top 10, reflecting differences in impact and possibly healthcare responses.

Recovery percentages for all provinces/states in the analysis are very high (close to or above 98%), demonstrating substantial recovery among infected individuals.

The combined visualizations highlight differences in case counts, deaths, and recoveries across major states, offering a clear comparison of COVID-19’s impact and outcomes at the state/province level.
